```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# ===== 한글 폰트 설정 =====
font_path = "./NanumGothic.ttf"   # 같은 폴더에 NanumGothic.ttf 필요
fontprop = fm.FontProperties(fname=font_path)

# DataFrame 변환
df_distance = pd.DataFrame(distance)
df_action = pd.DataFrame(action_time)
df_total = pd.DataFrame(total_time)

# 합계 제거
df_distance = df_distance[df_distance["시도"] != "합계"]
df_action = df_action[df_action["시도"] != "합계"]
df_total = df_total[df_total["시도"] != "합계"]

target = "부산"

def compare_plot(df, title, exclude_cols=["시도","이송 인원"]):
    cols = [c for c in df.columns if c not in exclude_cols]
    
    # 구간별 비율 계산 (%)
    ratio_df = df.copy()
    for c in cols:
        ratio_df[c] = df[c] / df["이송 인원"] * 100
    
    # 부산
    busan = ratio_df[ratio_df["시도"] == target][cols].iloc[0]
    # 나머지 평균
    others = ratio_df[ratio_df["시도"] != target][cols].mean()
    
    # 막대그래프
    x = range(len(cols))
    plt.figure(figsize=(12,6))
    plt.bar(x, others, width=0.4, label="나머지 평균")
    plt.bar([i+0.4 for i in x], busan, width=0.4, label="부산")
    plt.xticks([i+0.2 for i in x], cols, rotation=45, fontproperties=fontprop)
    plt.title(title, fontproperties=fontprop)
    plt.ylabel("비율(%)", fontproperties=fontprop)
    plt.legend(prop=fontprop)
    plt.tight_layout()
    plt.show()

# 거리 기준
compare_plot(df_distance, "거리 기준(비율): 부산 vs 나머지 평균")

# 출동 소요시간 기준
compare_plot(df_action, "출동 소요시간 기준(비율): 부산 vs 나머지 평균")

# 총 소요시간 기준
compare_plot(df_total, "총 소요시간 기준(비율): 부산 vs 나머지 평균")
```
